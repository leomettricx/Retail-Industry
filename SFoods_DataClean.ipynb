#### TYype of Product (Family)

```python

def famProdutos(xCod):
    x = str(xCod)
    yCod = x[:2]
    if yCod == '01':
        return 'Blister'
    elif yCod =='02':
        return 'Temperos'
    elif yCod =='03':
        return 'Molhos'
    elif yCod == '04':
        return 'Conservas'
    elif yCod =='05': 
        return 'Institucional'
    elif yCod == '06':
        return 'Diversos '
    elif yCod =='07':
        return 'Insumos'
    elif yCod =='08':
        return 'Consumo'
    elif yCod == '09':
        return 'Produtos Prontos para Venda'
    elif yCod =='10':
        return 'Serviço'
    elif yCod == '11':
        return 'Embalagens'
    elif yCod =='12':
        return 'Contas Fixas'
    elif yCod == '13':
        return 'Bem Ativo'
    elif yCod =='14':
        return 'Manutenção de Ativo'
    elif yCod =='15':
        return 'Display'
    elif yCod == '16':
        return 'Supra Life'

```
