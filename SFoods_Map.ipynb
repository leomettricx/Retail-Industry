```python
import pandas as pd
import numpy as np
```

```python
info = []
i=1
data= open('MunicipiosBrasil.csv')
for line in data:
    info.append(line.rstrip('\n').split(';')) 
       
cidades = pd.DataFrame(data = info, columns=['ID','LATITUDE','LONGITUDE','cidade','UF','Valor']).reset_index(drop=True)   
data.close()
```

```python
cidades.drop(['ID'],axis = 1, inplace = True) # s1.drop(['Valor']) # , axis = 1, inplace = True
```
