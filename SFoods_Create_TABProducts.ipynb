```python

file = open('supraProdutos.csv','w')
file.write('cod'+','+'descricao'+'\n') 
a1 = sheet.cell(row=i, column=1)
codigo = []
descricao = []

j = 0
e = sheet.cell(row=i, column=2).value 
    
while i < 7580:
    if sheet.cell(row=i, column=1).value != None:
        a2 = sheet.cell(row=i, column=2)
        a1 = sheet.cell(row=i, column=1)
        e = sheet.cell(row=i, column=2).value 
      
   
        codigo.append(a1.value)
        descricao.append(a2.value)
     

    i += 1

 

    while sheet.cell(row=i, column=1).value == None:
                
                 
        if e is not None:
            descricao[j] = e + ' '+str(sheet.cell(row=i, column=2).value)
            e = e + ' '+ str(sheet.cell(row=i, column=2).value)
               
        i += 1

    print(j,' Codigo  : ',codigo[j])
    print('Descricao : ',descricao[j])
    print(i)
    de = descricao[j]
    de = de.replace(',','.')
    
  
  #produto = {
  #   "codigo":codigo[j],
  #   "descricao":descricao[j]
  # }
  #rec_id = collection.insert_one(produto)
    file.write(str(codigo[j]).zfill(6)+','+de+'\n') 
  

    j += 1  

file.close()
```
