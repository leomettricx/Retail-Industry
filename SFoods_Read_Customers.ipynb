```python
clientes = openpyxl.load_workbook('cli.xlsx')
```

```python
sheet = clientes.active
```

```python
i=2
```

```python

a1 = sheet.cell(row=i, column=1)
empresa = []
endereco = []
fantasia = []
j = 0

while i < 9608:
  if sheet.cell(row=i, column=1).value != None:
    a2 = sheet.cell(row=i, column=3)
    a3 = sheet.cell(row=i, column=5)
    a4 = sheet.cell(row=i, column=6)
    a5 = sheet.cell(row=i, column=12)
     
    if a5 is None:
        revenda = 'SIM'
    else:
        revenda = a5.value
  
    empresa.append(a2.value)
    fantasia.append(a3.value)
    endereco.append(a4.value)
   
   
  i += 1
  e = a4.value
  f = a3.value
    

  while sheet.cell(row=i, column=1).value == None:
                
        if a2 is not None and sheet.cell(row=i, column=3).value is not None:
           empresa[j] =(a2.value + str(sheet.cell(row=i, column=3).value))
          
        if e is not None and sheet.cell(row=i, column=6).value is not None:
           endereco[j] = e + ' '+str(sheet.cell(row=i, column=6).value)
           e = e + ' '+ str(sheet.cell(row=i, column=6).value)
        else:
           endereco[j] = ''
           e = e + ' '
            
        if f is not None and sheet.cell(row=i, column=5).value is not None: 
            fantasia[j] = f + sheet.cell(row=i, column=5).value
            f = f + str(sheet.cell(row=i, column=5).value)
       
        i += 1
 
  ex = e.strip()
  print(j,' Empresa  : ',empresa[j])
  print('Fantasia : ',fantasia[j])
  primeiro = e.find("-")
  segundo  = e.find('-', primeiro+1)

  
  address= ex[0:primeiro-1].strip() 
  bairro = ex[primeiro+1:segundo].strip()
  cidade = ex[segundo+2:len(ex)-5].strip()
  estado = ex[-2:]
    
  print('Endereço : ',address)
  print('Bairro   : ',bairro)
  print('Cidade   : ',cidade)
  print('Estado   : ',estado)
  print('Revenda  : ',revenda)

 
  print()
  cliente = {
     "Empresa":empresa[j],
     "NomeFantasia":fantasia[j],
     "Endereco":address,
     "Bairro":bairro,
     "Cidade":cidade,
     "Estado":estado    
  }
  #rec_id = collection.insert_one(cliente)
  

  j += 1  

```
